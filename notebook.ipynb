```python
import pandas as pd

dataset = pd.read_csv('news_200.csv')
```

```python
news_title = dataset['title']

print(news_title)
```

```python
import spacy

nlp = spacy.load("en_core_web_sm")

news_article = dataset['title']

processed_texts = []
for entry in news_article:
    title = entry
    doc = nlp(title)
    print(doc)
    processed_texts.append(doc)
```

```python

triples = []

for text in processed_texts:
    for token in text:
        #print(text, token, token.dep_, token.pos_)

        if token.dep_ == "nsubj" and token.head.pos_ == "VERB":
            subject = token.text
            verb = token.head.lemma_
            for child in token.head.children:
                if child.dep_ == "dobj":
                    obj = child.text
                    triples.append((subject, verb, obj, text))  # Store the text along with the triple

```

```python

# Print the triples and the text they were extracted from
for triple in triples:
    subject, verb, obj, text = triple
    print("Text:", text)
    print("Tripla:",subject,verb,obj)
   
    print("=" * 50)
```

```python
for triple in triples:
    subject, verb, obj, text = triple
    subject_doc = nlp(subject)
    object_doc = nlp(obj)

    # Calculate similarity between subject and object
    similarity_score = subject_doc.similarity(object_doc)

    print("Triple:", subject, verb, obj)
    print("Text:", text)
    print("Similarity Score:", similarity_score)
    print("=" * 50)
```

```python
import spacy
from itertools import combinations

# Load the pre-trained model with word vectors
nlp = spacy.load("en_core_web_md")

# Assuming you have a list of triples like triples = [(subject, verb, obj, text), ...]
highest_similarity_pairs = []
highest_similarity_score = 0.0

for triple1, triple2 in combinations(triples, 2):
    subject1, verb1, obj1, _ = triple1
    subject2, verb2, obj2, _ = triple2

    # Process the subjects, verbs, and objects using spaCy
    subject1_doc = nlp(subject1)
    verb1_doc = nlp(verb1)
    obj1_doc = nlp(obj1)

    subject2_doc = nlp(subject2)
    verb2_doc = nlp(verb2)
    obj2_doc = nlp(obj2)

    # Calculate similarity between triples
    subject_similarity = subject1_doc.similarity(subject2_doc)
    verb_similarity = verb1_doc.similarity(verb2_doc)
    obj_similarity = obj1_doc.similarity(obj2_doc)

    # Calculate overall similarity score (average of subject, verb, and object similarity)
    overall_similarity = (subject_similarity + verb_similarity + obj_similarity) / 3

    if overall_similarity > highest_similarity_score:
        highest_similarity_score = overall_similarity
        highest_similarity_pairs = [(triple1, triple2)]
    elif overall_similarity == highest_similarity_score:
        highest_similarity_pairs.append((triple1, triple2))

# Print the pairs with the highest similarity scores
for pair in highest_similarity_pairs:
    print("Triple 1:", pair[0])
    print("Triple 2:", pair[1])
    print("Overall Similarity Score:", highest_similarity_score)
    print("=" * 50)
```
